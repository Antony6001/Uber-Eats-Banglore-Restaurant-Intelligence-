```python
for col in df.columns:
    print(f"{col}: {df[col].dtype}")
```

```python
df['name'] = df['name'].str.encode('latin1', errors='ignore').str.decode('utf-8', errors='ignore')
```

```python
# for removing unwanted charecters in name column

import re

df['name'] = df['name'].apply(lambda x: re.sub(r'[^a-zA-Z0-9 &]', '', x))
```

```python
# converting the uppercase to lower_case

df.name=df.name.str.lower().str.replace('_',' ')
```

```python
df.online_order=df.online_order.str.lower().str.replace(' ','_')
```

```python
df.book_table.value_counts()
```

```python
df.book_table=df.book_table.str.lower()
```

```python
import ast
```

```python
# convert the NEW (str) to numeric

df['rate']=df['rate'].replace(['NEW',None])
```

```python
#Removing the "/5" , and dtype is changed here ,

df['rate']=df['rate'].str.replace('/5','').astype(float)
```

```python
df.votes.dtype
```

```python
df.phone=df.phone.str.replace(r'\D','',regex=True)
```

```python
df['mobile'] = df['phone'].str[-10:]
```

```python
df['landline'] = df['phone'].str[:-10]
```

```python
df['landline'] = df['landline'].str.replace('^91', '', regex=True)
```

```python
df['mobile'] = df['phone'].str.extract(r'(\d{10})$')
```

```python

df.mobile=df.mobile.astype(int)
```

```python
df['landline'] = pd.to_numeric(df['landline'], errors='coerce')
```

```python
# upper case to convert the lower case

df.location=df.location.str.lower().str.replace(' ','_')
```
