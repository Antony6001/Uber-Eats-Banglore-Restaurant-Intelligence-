#Data cleaning

Missing values find and handle

```python
for col in df.columns:
  miss=df[col].isnull().sum()
  print(f"{col}:{miss}")
```

```python
df.restaurant_name
```

```python
df.restaurant_name = df.restaurant_name.str.lower()
```

```python
df.restaurant_name=df.restaurant_name.str.strip()
```

```python
df.restaurant_name=df.restaurant_name.str.replace("-","")
```

```python
df['restaurant_name'] = df['restaurant_name'].str.replace(r'[^a-zA-Z\s]', '', regex=True)
```

```python
df.order_date
```

```python
import datetime
```

```python
df['order_date']=pd.to_datetime(df['order_date'],errors='coerce')
```
